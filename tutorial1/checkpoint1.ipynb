<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
df['INSTRUCTION_EXPENDITURE_PER_STUDENT'] = df['INSTRUCTION_EXPENDITURE'] / df['GRADES_ALL_G']
df['PORTION_OF_REVENUE_USED'] = df['INSTRUCTION_EXPENDITURE'] / df['TOTAL_REVENUE']
df.dropna(subset=['AVG_MATH_8_SCORE'], inplace=True)
```

Feature engineering justification: **I created an instruction expenditure per student column to see if how much money spent per student has any effect on academic performance. I also created a column for how much of a state's total revenue was being spent on instruction expenditure to see if states that seemed to prioritize funding instruction had better academic performances. Lastly, I dropped any missing 8th grade Math Scores since those won't contribute to my data visualization.**

<h2>Visualization</h2>

Investigate the relationship between your chosen response variable and at least two predictors using visualizations. Write down your observations.

**Visualization 1**

```python
df.plot.scatter(x='INSTRUCTION_EXPENDITURE_PER_STUDENT', y='AVG_MATH_8_SCORE', alpha=0.6)
plt.xlabel('Instruction Expenditure Per Student')
plt.ylabel('8th Grade Math Score')
```

**As you can see from the scatter plot, it would appear that having a higher instruction expenditure per student has a positive correlation with improved Math 8 Scores and academic success, though at $6/student the benefit begins to plateau a bit.**

**Visualization 2**

```python
df.plot.scatter(x='PORTION_OF_REVENUE_USED', y='AVG_MATH_8_SCORE', alpha=0.6)
plt.xlabel('Portion of Revenue Used On Instruction Expenditure')
plt.ylabel('8th Grade Math Score')
```

**From this scatter plot it seems like there is a somewhat weak positive correlation between portion of revenue used on instruction expenditure and 8th Grade Math Scores.**

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['INSTRUCTION_EXPENDITURE_PER_STUDENT','PORTION_OF_REVENUE_USED']].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']

y.fillna(y.median(), inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
  X, y, test_size=.3, random_state=42)

```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.intercept_)
print(model.coef_)

print(model.score(X_test, y_test))
np.mean(model.predict(X_test)-y_test)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'INSTRUCTION_EXPENDITURE_PER_STUDENT'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('8th Grade Math Scores')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'PORTION_OF_REVENUE_USED'

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('8th Grade Math Scores')
plt.title("Model Behavior on Testing Set")
```
